```r
library(httr)
library(readxl)
library(stringr)

# Read the Excel
data <- read_excel("Rosmarinus officinalis (ligands) 1.xlsx")

# Create a folder to store the downloaded SDFs
dir.create("sdf_Rosmarinus officinalis", showWarnings = FALSE)

download_sdf <- function(compound_name) {
  name_encoded <- URLencode(compound_name)

  # Construct the URL to retrieve the sdf  from PubChem
  url <- paste0("https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/",
                name_encoded, "/SDF")

  response <- GET(url)

  if (response$status_code == 200) {
    # Extract the SDF content as text
    sdf_content <- content(response, as = "text", encoding = "UTF-8")

    # Create a safe filename by replacing non-alphanumeric characters
    file_name <- paste0("sdf_Rosmarinus officinalis/", str_replace_all(compound_name, "[^[:alnum:]]+", "_"), ".sdf")

    writeLines(sdf_content, file_name)

    message("Downloaded: ", file_name)
  } else {
    message("Failed to download SDF for: ", compound_name)
  }
}

# Loop for each phytochemical name in  excel and download its SDF file
lapply(data$`Phytochemical name`, download_sdf)
```
